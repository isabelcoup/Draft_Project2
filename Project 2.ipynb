## Introduction

##### Scientific question: What genetic changes in the auditory processing pathway of mammals allow for high-frequency hearing, leading to the convergent evolution of echolocation?

Echolocation is the location of objects using reflected sound. Three species of mammals that use echolocation are Eptesicus fuscus (Big brown bat), Hipposideros armiger (Great roundleaf bat), and Orcinus orca (Killer whale). Three mammals that cannot echolocate are Cynopterus sphinx (Greater short-nosed fruit bat), Pteropus vampyrus (Large flying fox), and Homo sapiens. 

Some genes known to be involved with the auditory pathway and hearing loss include: 
- Prestin (SLC26A5; involved in cochlear hair cell motility and ion exchange, critical for hearing sensitivity), 
- GJB2 (associated with hearing loss in humans), 
- CDH23 (responsible for stereocilia organization and hair bundle formation), 
- PCDH15 (maintenance of normal retinal and cochlear function), 
- OTOF (vesicle membrane fusion, associated with deafness in mice), and 
- KCNQ4 (cochlear neuron excitability).

The data used in this project was sourced from NCBI Nucleotide and NCBI GEO:
- NCBI Nucleotide https://www.ncbi.nlm.nih.gov/nuccore As stated on this databases website, "The Nucleotide database is a collection of sequences from several sources, including GenBank, RefSeq, TPA and PDB. Genome, gene and transcript sequence data provide the foundation for biomedical research and discovery."
- NCBI GEO https://www.ncbi.nlm.nih.gov/geo/ As stated on this website, "GEO is a public functional genomics data repository supporting MIAME-compliant data submissions. Array- and sequence-based data are accepted. Tools are provided to help users query and download experiments and curated gene expression profiles."

##### Scientific hypothesis: If there are similar changes in the sequences and expression of hearing-related genes* in the brain of mammals, then an auditory processing pathway will evolve, allowing for the use of echolocation. 

Coding sequences of each gene was downloaded as a FASTA file from NCBI nucleotide. The sequences for each species was then compiled into one file for each gene and Multiple sequence alignment was performed using Clustalw2. A phylogenetic tree was created for the alignment of each gene. 
- Multiple sequence alignment allows me to compare the sequences of a gene for each species, then plot it using a phylogenetic tree to determine which gene(s) are most similar among echolocating species. 

RNA-Seq gene expression data is from NCBI GEO, samples come from the brains of 5 echolocating and 3 non-echolocating species of bats. https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE40993 The data for all 8 organisms was compiled into one file and run through RNA-Seq via BioJupies. 
- RNA-seq allows us to compare the gene expression data for each species, then plot the logF2C and p-value on a volcano plot to see which genes are most differentially expressed between echolocating and non-echolocating species.

# Analysis 1: Multiple sequence alignment using ClustalW2

Multiple sequence alignment is the alignment of three or more biological sequences of similar lengths. ClustalW2 produces multiple sequence alignments of divergent sequences by calculating the best match for the selected sequences and lining them up so that the identities, similarities and differences can be seen. The file input used in this project is a FASTA file containing the sequences of a gene for each species, although multiple file types are accepted. The output is a CLUSTAL file that shows the alignments. It can also output a guide tree file (.dnd) and phylogenetic tree (.ph). Evolutionary relationships can be seen by viewing the phylogenetic tree. Species that are clustered closest together have the most similar sequences-- they are closest related. 


Clustalw2 was downloaded (http://www.clustal.org/download/current/) and performed in the computer terminal, as it was designed for interactive use and doesn't perform as well in jupyter notebook (I was unable to download the files through Jupyter notebook). 
- First, open terminal and type clustalw2
- From the list, select option (1) Sequence Input From Disc
- Type in file name [prestin.txt]
- Select option (2) Multiple Alignments
- Next, select option (1) Do complete multiple alignment now Slow/Accurate
- Type in name for CLUSTAL output file [prestin.aln], pairwise alignment will be performed and guide tree will be established. 
- Type in name for new guide tree [prestin.dnd]
- Multiple sequence alignment will be performed based on guide tree and alignment file will be exported.
    - Based on guide tree, closest sequences are compared first, leaving the most unrelated sequence for comparison last. 
- Back at the main menu, select option (4) Phylogenetic trees
- Next, select option (4) Draw tree now
- Type in name for PHYLIP tree output file [prestin.ph]
- Analysis is complete and you may leave program

# clustalw2

```python
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
from Bio.Phylo.TreeConstruction import DistanceCalculator
from Bio import AlignIO
aln = AlignIO.read(open('alignedp.aln'), 'clustal')
constructor = DistanceTreeConstructor()
calculator = DistanceCalculator('identity')
dm = calculator.get_distance(aln)
upgmatree = constructor.upgma(dm)
print(upgmatree)

njtree = constructor.nj(dm)
print(njtree)

Phylo.draw(upgmatree)

Phylo.draw(njtree)

scorer = Phylo.TreeConstruction.ParsimonyScorer()
searcher = Phylo.TreeConstruction.NNITreeSearcher(scorer)
constructor = Phylo.TreeConstruction.ParsimonyTreeConstructor(searcher, njtree)
pars_tree = constructor.build_tree(aln)
print(pars_tree)
Phylo.draw(pars_tree)
```
